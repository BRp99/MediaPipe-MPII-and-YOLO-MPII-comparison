```python
from ultralytics import YOLO
import os
from scipy.io import loadmat
import numpy as np
import shutil
import math
import scipy.io
#########################
import matplotlib.pyplot as plt
from tqdm import tqdm
#from adjustText import adjust_text
from matplotlib.patches import Circle
import copy
from pathlib import Path
import sys
```

```python
# Load model YOLO v8
model = YOLO("yolov8m-pose.pt")

# Configurations
N = 10
MPII_IMAGES_DIR = "./mpii/images"
yolo_images_dir = "./YOLO_images"

# Load annotations MPII
annotations_data = loadmat("./mpii/annotations/mpii_human_pose.mat")
annolist_test = annotations_data['RELEASE'][0,0]['annolist'][0]

# Index to joint name conversion
mpii_idx_to_jnt = {0: 'rankl', 1: 'rknee', 2: 'rhip', 5: 'lankl', 4: 'lknee', 3: 'lhip',
                   6: 'pelvis', 7: 'thorax', 8: 'upper_neck', 11: 'relb', 10: 'rwri', 9: 'head',
                   12: 'rsho', 13: 'lsho', 14: 'lelb', 15: 'lwri'}

# Template to update as and when ground truth is read
mpii_template = dict([(mpii_idx_to_jnt[i], []) for i in range(16)])

# Initialize counter
count = 0
results_list = []

# Iterate over images in MPII dataset
for annolist in annolist_test:
    image_name = annolist['image']['name'][0, 0][0]
    image_file = os.path.join(MPII_IMAGES_DIR, image_name)
    shutil.copy(image_file, yolo_images_dir)
    
    # Execute detection for YOLO_images directory
    results = model(source=image_file, show=False, conf=0.3, save=False)
    results_list.append(results)
    
    count += 1
    if count >= N:
        break

# Iterate over detection results
for i, result in enumerate(results_list):
    for r in result:
        image_name = r.path.split('\\')[-1]
        xyn = r.keypoints.xyn
        for j, pose in enumerate(xyn):
            for k, landmark in enumerate(pose):
                x, y = landmark
                print(f"Image {i+1} ({image_name}) - Pose {j+1} - Keypoint {k+1}:\t ({x}, {y})")
```
